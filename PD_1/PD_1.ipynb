Задание 1

Скачайте с сайта grouplens.org...movielens/ датасет любого размера.

Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
mv = pd.read_csv('movies.csv')
mv.head()
```

```python
rt = pd.read_csv('ratings.csv')
rt.head()
```

```python
# Объединил 2 csv обычным джойном

mv_rt = mv.merge(rt, how='inner')
mv_rt
```

```python
#  Сделал сначала сортировку по рейтингу, сбросил индексы, затем к колонке title применил подсчет количества и получил первый title фильма

mv_rt[mv_rt.rating == 5.0].reset_index()['title'].value_counts().head(1)
```

Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года.
 
Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
power['country'].unique()
```

```python
# Фильтрую страны
filtered_country = power[(power['country'] == 'Estonia') | (power['country'] == 'Latvia') | (power['country'] == 'Lithuania')]
filtered_country
```

```python
# Фильтрую категории
filtered_country_category = filtered_country[ (filtered_country['category'] == 4) | (filtered_country['category'] == 12) | (filtered_country['category'] == 21)]
filtered_country_category
```

```python
# Фильтрую года и quantity, проверяю что бы не было отрицательных значений сортировкой
filtered_country_category_year = filtered_country_category[ (filtered_country_category['year'] >= 2004) & (filtered_country_category['year'] <= 2010) ]
filtered_country_category_year = filtered_country_category_year[ (filtered_country_category_year['quantity'] >= 0 ) ]
filtered_country_category_year.sort_values('quantity')
```
