Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

    если источник traffic_source равен yandex или google, то в source_type ставится organic
    для источников paid и email из России - ставим ad
    для источников paid и email не из России - ставим other
    все остальные варианты берем из traffic_source без изменений


```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
# Вариант 1
def check_source_type(df):
    if df['traffic_source'] == 'yandex' or df['traffic_source'] == 'google':
        return 'organic'
    elif df['region'] == 'Russia' and (df['traffic_source'] == 'paid' or df['traffic_source'] == 'email'):
        return 'ad'
    elif df['region'] != 'Russia' and (df['traffic_source'] == 'paid' or df['traffic_source'] == 'email'):
        return 'other'
    else:
        return df['traffic_source']
```

```python
# Вариант 2
log.loc[(log['traffic_source'] == 'yandex') | (log['traffic_source'] == 'google'), 'source_type'] = 'organic'
log.loc[(log['region'] == 'Russia') & (log['traffic_source'] == 'paid') | (log['traffic_source'] == 'email'), 'source_type'] = 'ad'
log.loc[(log['region'] != 'Russia') & (log['traffic_source'] == 'paid') | (log['traffic_source'] == 'email'), 'source_type'] = 'other'
```

```python
log.loc[log['source_type'].isnull(), 'source_type'] = log['traffic_source']
log.head(30)
```

```python
log['source_type'] = log.apply(check_source_type, axis=1)
log.head(20)
```

Задание 3

    Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
    Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
    Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.


```python
df = pd.read_csv('ratings.csv')
df
```

```python
# Группирую по id пользователя и считаю количество, это и будет количеством оценок
grouped = df.groupby('userId').count().reset_index()
# Фильтрую по количеству в столбце рейтинга больше 100
grouped = grouped[grouped['rating'] > 100]
# Удаляю ненужные столбцы
grouped.drop(['movieId', 'rating', 'timestamp'], axis=1, inplace=True)
grouped
```

```python
# Объединяю полученный датафрейм только с id юзеров, которые проставили 
# больше 100 оценок с первоначальным датафреймом, что бы получить время простановки оценок
merged = grouped.merge(df, on='userId', how='inner')
merged
```

```python
def diff_(df):
    return df['timestamp']['max'] - df['timestamp']['min']
```

```python
# Группирую объединенный датафрейм по столбцу timestamp с agg функциями min и max
# Что бы получить минимальное и максимальное значение для каждого пользователя
merged = merged.groupby('userId').agg({'timestamp': ['min', 'max']}).reset_index()
merged
```

```python
merged['diff'] = merged.apply(diff_, axis=1)
merged
```

```python
# Считаю среднее по столбцу в часах
merged['diff'].mean() / 60 / 60
```
